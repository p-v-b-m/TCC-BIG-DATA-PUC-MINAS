# Importando dataset com as taxas de câmbio (dólar americano)
https://dados.gov.br/dataset/10813-taxa-de-cambio-livre-dolar-americano-compra/resource/f02d92bd-bf7d-4bfe-b1c3-06938f7012c2

```python
txcambio=pd.read_csv("bcdata.sgs.10813.csv", sep=";")
```

```python
txcambio.head()
```

```python
txcambio['data'] =  pd.to_datetime(txcambio['data'], format='%d/%m/%Y')
```

```python
txcambio['ANO-MES'] = txcambio['data'].apply(lambda x: x.strftime('%Y-%m')) 
```

```python
txcambio.head(5)
```

```python
del txcambio["data"]
txcambio.head()
```

```python
txcambio=txcambio[["ANO-MES","valor"]]
txcambio
```

```python
txcambio.valor.replace(",",".",regex=True, inplace=True)
```

```python
txcambio["valor"]=txcambio["valor"].astype("Float64")
```

```python
txcambiomensal=txcambio.groupby("ANO-MES").mean()
```

```python
txcambiomensal
```

```python
txcambiomensal["índice"]=pd.Series(np.arange(1,437), index=txcambiomensal.index)
```

```python
txcambiomensal=txcambiomensal.reset_index().set_index("índice",drop=True)
```

```python
txcambiomensal=txcambiomensal[["ANO-MES","valor"]]
```

```python
txcambiomensal["câmbio_mês_anterior"]=np.zeros(436)    
```

```python
txcambiomensal
```

```python
x = np.arange(2,437)
txcambiomensal["valor"].astype(float)
txcambiomensal["câmbio_mês_anterior"].astype(float)
for i in x:
    #print(txcambiomensal.loc[[i],"ANO-MES"])
    #print(txcambiomensal.loc[[i-1],"valor"])
    #txcambiomensal.loc[[i],"câmbio_mês_anterior"]=txcambiomensal.loc[[i-1],"valor"]
    txcambiomensal["câmbio_mês_anterior"]=txcambiomensal["valor"].shift(1)
```

```python
txcambiomensal
```

```python
txcambiomensal.rename(columns={'valor': 'câmbio_mês_atual'}, inplace = True)
```

```python
txcambiomensal
```
